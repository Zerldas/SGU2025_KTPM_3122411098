```python
%%bash
# Số thứ nhất là số lớn nhất
case1=(3 2 1)
# Số thứ hai là số lớn thứ hai
case2=(1 3 2)
# Số thứ ba là số lớn thứ ba
case3=(1 2 3)
# Có tất cả là số âm
case4=(-1 2 -3)
# Có ít nhất 2 số giống nhau
case5=(2 1 2)
# Cả 3 số đều giống nhau
case6=(1 1 1)

{
    echo "===========Kết quả cho test case của câu e========="

    echo "Case thứ nhất: Số thứ nhất là số lớn nhất"
    echo "Input: ${case1[*]}"
    echo "Output:"
    echo ${case1[*]} | ./e

    echo "================================="

    echo "Case thứ hai: Số thứ hai là số lớn thứ hai"
    echo "Input: ${case2[*]}"
    echo "Output:"
    echo ${case2[*]} | ./e

    echo "================================="

    echo "Case thứ ba: Số thứ ba là số lớn thứ ba"
    echo "Input: ${case3[*]}"
    echo "Output:"
    echo ${case3[*]} | ./e

    echo "================================="

    echo "Case thứ tư: Có tồn tại số âm"
    echo "Input: ${case4[*]}"
    echo "Output:"
    echo ${case4[*]} | ./e

    echo "================================="

    echo "Case thứ năm: Có tồn tại ít nhất hai số giống nhau"
    echo "Input: ${case5[*]}"
    echo "Output:"
    echo ${case5[*]} | ./e

    echo "================================="

    echo "Case thứ sáu: Có tồn tại ít nhất hai số giống nhau"
    echo "Input: ${case6[*]}"
    echo "Output:"
    echo ${case6[*]} | ./e
} >> result.txt
```

```python
%%bash

# a = 0, b = 0, c = 0 => Vô số ngiệm
case1=(0 0 0)
# a = 0, b = 0, => Vô nghiệm
case2=(0 0 2)
# a = 0 => Nếu -c / b < 0 => Vô nghiệm Hoặc -c / b > 0 => Pt có 2 nghiệm
# Case -c / b  < 0
case3=(0 1 2)
# Case -c / b > 0
case4=(0 2 -10)
# Case delta < 0
case5=(1 0 1)
# Case delta = 0 => Nếu y < 0 => Vô nghiệm Hoặc đặc biệt nếu y >= 0 thì pt có thể có 2 nghiệm
# Case delta = 0, y < 0 => Pt vô nghiệm
case6=(1 2 1)
# Case delta = 0, y >=0 => Pt có 2 nghiệm
case7=(1 -2 1)
# Case delta > 0
# Case cả 2 nghiệm y1 và y2 điều dương => Pt có 4 nghiệm
case8=(1 -3 2)
# Case 1 nghiệm âm và 1 nghiệm dương => Pt có 2 nghiệm
case9=(1 1 -2)
# Case cả 2 nghiệm điều âm => Pt vô nghiệm
case10=(1 3 2)

{
    echo "========Kết quả test cho hàm solveQuartic=========="

    echo "Case thứ nhất: a = 0, b = 0, c = 0"
    echo "Input: ${case1[*]}"
    echo "Output:"
    echo ${case1[*]} | ./solveQuartic

    echo "================================="

    echo "Case thứ hai: a = 0, b = 0"
    echo "Input: ${case2[*]}"
    echo "Output:"
    echo ${case2[*]} | ./solveQuartic

    echo "================================="

    echo "Case thứ ba: a = 0, y < 0"
    echo "Input: ${case3[*]}"
    echo "Output:"
    echo ${case3[*]} | ./solveQuartic

    echo "================================="

    echo "Case thứ tư: a = 0, y > 0"
    echo "Input: ${case4[*]}"
    echo "Output:"
    echo ${case4[*]} | ./solveQuartic

    echo "================================="

    echo "Case thứ năm: delta < 0"
    echo "Input: ${case5[*]}"
    echo "Output:"
    echo ${case5[*]} | ./solveQuartic

    echo "================================="

    echo "Case thứ sáu: delta = 0, y < 0"
    echo "Input: ${case6[*]}"
    echo "Output:"
    echo ${case6[*]} | ./solveQuartic 

    echo "================================="

    echo "Case thứ sáu: delta = 0, y > 0"
    echo "Input: ${case7[*]}"
    echo "Output:"
    echo ${case7[*]} | ./solveQuartic 

    echo "================================="

    echo "Case thứ sáu: delta > 0, y1 > 0 && y2 > 0"
    echo "Input: ${case8[*]}"
    echo "Output:"
    echo ${case8[*]} | ./solveQuartic 

    echo "================================="

    echo "Case thứ sáu: delta > 0, (y1 < 0 && y2 > 0) || (y1 > 0 && y2 < 0)"
    echo "Input: ${case9[*]}"
    echo "Output:"
    echo ${case9[*]} | ./solveQuartic 

    echo "================================="

    echo "Case thứ sáu: delta > 0, y1 < 0 && y2 < 0"
    echo "Input: ${case10[*]}"
    echo "Output:"
    echo ${case10[*]} | ./solveQuartic
} >> result.txt
```
